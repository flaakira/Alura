# Aula 1

Fazendo primeiras visualizacoes

```
import pandas as pd
```

```
df = pd.read_csv('monitoramento_tempo.csv')
```

```
df.head(5)
```

```
df.info()
```

```
df['data'] = pd.to_datetime(df['data'])
```

```
import matplotlib.pyplot as plt
%matplotlib inline
```

```
plt.plot(df['data'], df['temperatura'])
```

```
plt.figure(figsize=(15,6))
plt.plot(df['data'], df['temperatura'])
plt.title('Temperatura no tempo')
```

```
fig = plt.figure(figsize=(15,6))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'], color = 'lightgreen')

eixo.set_title('Temperatura no momento', fontsize = 25)
eixo.set_ylabel('Temperatura', fontsize = 20)
eixo.set_xlabel('Data', fontsize = 20)
eixo.legend(['Temperatura'], loc = 'lower left', fontsize = 15)
```

# Aula 2

Customizando nossas visualizacoes

```
fig = plt.figure(figsize=(15,6))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'], color = 'darkgreen', ls = 'dotted', lw = 2, marker = '.')

eixo.set_title('Temperatura no momento', fontsize = 25)
eixo.set_ylabel('Temperatura', fontsize = 20)
eixo.set_xlabel('Data', fontsize = 20)
eixo.legend(['Temperatura'], loc = 'lower left', fontsize = 15)
```

```
import datetime
```

```
fig = plt.figure(figsize=(15,6))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'], color = 'yellow')

eixo.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))
eixo.set_title('Temperatura no momento', fontsize = 25)
eixo.set_ylabel('Temperatura', fontsize = 20)
eixo.set_xlabel('Data', fontsize = 20)
eixo.legend(['Temperatura'], loc = 'lower left', fontsize = 15)
eixo.grid(True)
```

```
fig = plt.figure(figsize=(15,6))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'], color = 'yellow')

eixo.set_xlim(datetime.datetime(2015,1,1), datetime.datetime(2016,1,1))
eixo.set_title('Temperatura no momento', fontsize = 25)
eixo.set_ylabel('Temperatura', fontsize = 20)
eixo.set_xlabel('Data', fontsize = 20)
eixo.legend(['Temperatura'], loc = 'lower left', fontsize = 15)
eixo.grid(True)
```

```
fig = plt.figure(figsize=(15,6))
eixo = fig.add_axes([0,0,1,1])
eixo2 = fig.add_axes([0.1,0.65,0.3,0.3])
eixo3 = fig.add_axes([0.65,0.65,0.3,0.3])
eixo4 = fig.add_axes([0.1,0.1,0.3,0.3])
eixo5 = fig.add_axes([0.65,0.1,0.3,0.3])

eixo.grid(True)
eixo.plot(df['data'], df['temperatura'], color = 'grey')
eixo.set_title('Temperatura no momento', fontsize = 25, pad = 20)
eixo.set_ylabel('Temperatura', fontsize = 20)
eixo.set_xlabel('Data', fontsize = 20)
eixo.legend(['Temperatura'], loc = 'lower right', fontsize = 15)

eixo2.grid(True)
eixo2.set_xlim(datetime.datetime(2013,1,1), datetime.datetime(2014,1,1))
eixo2.plot(df['data'], df['temperatura'], color = 'blue')
eixo2.set_title('Temperatura em 2013', fontsize = 15)
eixo2.legend(['Temperatura'], loc = 'lower center', fontsize = 8)

eixo3.grid(True)
eixo3.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))
eixo3.plot(df['data'], df['temperatura'], color = 'green')
eixo3.set_title('Temperatura em 2014', fontsize = 15)
eixo3.legend(['Temperatura'], loc = 'lower center', fontsize = 8)

eixo4.grid(True)
eixo4.set_xlim(datetime.datetime(2015,1,1), datetime.datetime(2016,1,1))
eixo4.plot(df['data'], df['temperatura'], color = 'red')
eixo4.set_title('Temperatura em 2015', fontsize = 15)
eixo4.legend(['Temperatura'], loc = 'lower center', fontsize = 8)

eixo5.grid(True)
eixo5.set_xlim(datetime.datetime(2016,1,1), datetime.datetime(2017,1,1))
eixo5.plot(df['data'], df['temperatura'], color = 'yellow')
eixo5.set_title('Temperatura em 2016', fontsize = 15)
eixo5.legend(['Temperatura'], loc = 'lower center', fontsize = 8)
```

```
fig = plt.figure(figsize=(15,6))
eixo = fig.add_axes([0,0,1,1])
eixo2 = fig.add_axes([0.7,0.65,0.3,0.3])

eixo.grid(True)
eixo.plot(df['data'], df['temperatura'], color = 'green')
eixo.set_xlim(datetime.datetime(2014,5,1), datetime.datetime(2014,6,1))
eixo.set_ylim(270,320)
eixo.set_title('Temperatura no momento', fontsize = 25, pad = 20)
eixo.set_ylabel('Temperatura', fontsize = 20)
eixo.set_xlabel('Data', fontsize = 20)
eixo.legend(['Temperatura'], loc = 'lower right', fontsize = 15)

azul_esquerda = df['data'] < datetime.datetime(2014,5,1)
azul_direita = df['data'] > datetime.datetime(2014,6,1)

eixo2.grid(True)
eixo2.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))
eixo2.plot(df[azul_esquerda]['data'], df[azul_esquerda]['temperatura'], color='green')
eixo2.plot(df[azul_direita]['data'], df[azul_direita]['temperatura'],color='green')
eixo2.plot(df['data'], df['temperatura'], color = 'blue')
eixo2.set_title('Temperatura em 2013', fontsize = 15)
eixo2.legend(['Temperatura'], loc = 'lower center', fontsize = 8)
```
